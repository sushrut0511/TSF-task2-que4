```python
# Drop columns which dont give very much information
data=data.drop(['Postal Code'],axis=1)
```

```python
# finding corelation between variables
corelation=data.corr()
```

```python
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns,annot=True)
```

```python
# Finding  relation between shipping mode and profit
sns.barplot(x='Ship Mode',y='Profit',data=data)
```

```python
# Finding relation between region and profit

sns.stripplot(x="Region", y="Profit", data=data,jitter=True,palette='Set1',dodge=True)
```

```python
# relation between quantitative variables on a sctter plot
g = sns.PairGrid(data)
g.map(plt.scatter)
```

```python
g = sns.FacetGrid(data=data,col='Segment',aspect=1.5,height=7)
g.map(plt.hist,'Sub-Category')
```

```python
# Region wise count of category
sns.countplot(x= data.Region, hue = data.Category,color='yellow')
```

```python
# Region wise count of Ship Mode
df_shipping=data.iloc[ : , 0 : 1]
df_shipping_rename=df_shipping.rename(columns= {'Ship Mode': "shipping"})
sns.countplot(x= data.Region, hue=df_shipping_rename.shipping ,color='yellow')

```
